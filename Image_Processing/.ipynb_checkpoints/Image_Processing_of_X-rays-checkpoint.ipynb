## **Image Segmentation by K-means Clustering Algorithm** 

```python
import numpy as np
from sklearn.cluster import KMeans
import cv2
import datetime
from matplotlib import pyplot as plt
import time
import keras
from keras.layers import *
from keras.models import *
from keras.preprocessing import image
```

```python
image1=cv2.imread("img001.jpg")
image2=cv2.imread("img002.jpg")
image3=cv2.imread("img003.jpg")
image=[image1,image2,image3]
reshaped=[0,0,0]
for i in range(0,3):
    reshaped[i] = image[i].reshape(image[i].shape[0] * image[i].shape[1], image[i].shape[2])
```

```python
numClusters=list(map(int,input("Enter the number of culsters for image 1,2 and 3 respectively: ").split(" ")))
```

```python
clustering=[0,0,0]
for i in range(0,3):
    kmeans = KMeans(n_clusters=numClusters[i], n_init=40, max_iter=500).fit(reshaped[i])
    clustering[i] = np.reshape(np.array(kmeans.labels_, dtype=np.uint8),
    (image[i].shape[0], image[i].shape[1]))
```

```python
sortedLabels=[[],[],[]]
for i in range(0,3):
    sortedLabels[i] = sorted([n for n in range(numClusters[i])],
        key=lambda x: -np.sum(clustering[i] == x))
```

```python
kmeansImage=[0,0,0]
concatImage=[[],[],[]]
for j in range(0,3):
    kmeansImage[j] = np.zeros(image[j].shape[:2], dtype=np.uint8)
    for i, label in enumerate(sortedLabels[j]):
        kmeansImage[j][ clustering[j] == label ] = int((255) / (numClusters[j] - 1)) * i
    concatImage[j] = np.concatenate((image[j],193 * np.ones((image[j].shape[0], int(0.0625 * image[j].shape[1]), 3), dtype=np.uint8),cv2.cvtColor(kmeansImage[j], cv2.COLOR_GRAY2BGR)), axis=1)
```

## **Image Compression by DWT_IDWT_For_Images_Compression**

```python
# This function is for making image square one 
def crop_image(gray_image):
  height, width = gray_image.shape[:2]
  if (height>= width):
    dim=width
  else:
    dim=height
  x=0
  y=0
  c_img=gray_image[x:dim,y:dim]  
  return c_img
```

```python
# This function is the main function to perform either h_0,h_1,g_0,g_1 or any 1-d filter , is row is a flag to determine 
# if it will be applied on rows or columns 
def img_filtering(gray_image,img_filter,is_row):
    if (~ is_row):
      gray_image=gray_image.T # in case we apply filter on columns
    f_img=np.zeros(gray_image.shape)
    l=img_filter.shape[0]//2
    r=gray_image.shape[0]
    c=gray_image.shape[1]
    d=l+1
    for i in range (r):
      for j in range (c):
        #if ( (j<l) | (j>(c-d)) ):
        if ( (j>l-1) & (j<(c-d+1)) ): 
          f_img[(i,j)]=np.dot(gray_image[i,slice(j-l,j+l+1)],img_filter)
        else:  
          f_img[(i,j)]=gray_image[(i,j)]
    if (~ is_row):
      f_img=f_img.T  # in case we apply filter on columns

  
    return f_img
```

```python
# Apply down sampling on rows or columns based on isrow flag value 
def down_sampling(f_img,isrow):
  if (isrow):
    ds_img=f_img[:,::2] # it means let all rows as they are and drop columns 
  else :
    ds_img=f_img[::2,:] # it means the reverse
  return ds_img
```

```python
# Apply down sampling on rows or columns based on isrow flag value 
def up_sampling(f_img,isrow):
  m,n = f_img.shape
  if (isrow):    
    us_img = np.zeros((m,2*n),dtype=f_img.dtype) # make the array wider by inserting zeros in rows 
    us_img[:,::2] = f_img
  else :
    us_img = np.zeros((2*m,n),dtype=f_img.dtype) # make the array more lengthy  by inserting zeros in columns
    us_img[::2,:] = f_img
    
  return us_img 
```

```python
# Apply quantization by dividing array on a scalar and then make floor(result).i.e integer devision I choose it to be like that
# instead of rounding to be more crude and results in less symbols 
def quantization (ds_img,q_num):
  q_img=ds_img//q_num # integer devision 
  dq_num=q_num

  return q_img ,dq_num # number to go to decoder
```

```python
# Apply dequantization by multiplying array with the same scalar used in devision in quantization process .
def dequantization (q_img,dq_num):
  dq_img=q_img*dq_num
  
  return dq_img
```

```python
# After quantization , we will be willing to make stream encoded by Entropy encoder (huffman) so we will pass its symbols 
#in that case unique numbers in image , along with repetition of each number to calculate probability , and finally data stream
def value_prob_stream(org_array):
    import numpy as np 
    elements,repetition=np.unique(org_array,return_index=False, return_inverse=False, return_counts=True, axis=None)
    o=list(elements)
    G=list(repetition)
    a=list(org_array)
    return o,G,a 
```
